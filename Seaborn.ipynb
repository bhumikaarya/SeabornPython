# SEABORN

***SEABORN IS A LIBRARY FOR MAKING STATISTICALGRAPH FOR VISUALIZATION.
ITS IS ADVANTAGE OVER MATPLOTLIB THAT IS IN MATPLOTLIB WE USED TO REPRESENT ONLY 2D FORM OF DATA BUT IN SEABORN LIBRARY WE CAN DRAW THE STATISTICAL GRAPH USING PYTHON.
ITS IS BUILT ON MATPLOTLIB AND PANDAS.***

# DIFFERENT CATEGORIED OF PLOT IN SEABORN:

***1)RELATIONAL PLOT
2)CATEGORICAL PLOT
3)DISTRIBUTION PLOT
4)REGRESSION PLOT
5)MATRIX PLOT
6)MULTI-PLOT GRIDS***

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var1=[1,2,3,4,5,6,7,8]
var2=[8,7,6,5,4,3,2,1]
#using matplotlib
plt.plot(var1,var2)
```

```python
#using seaborn
sns.lineplot(x=var1,y=var2)
plt.show()
```

```python
var1=[1,2,3,4,5,6,7,8]
var2=[8,7,6,5,4,3,2,1]

import pandas as pd
df=pd.DataFrame({'var1':var1,'var2':var2})
df
```

```python
#building graph using seaborn library
#lineplot is used to plot the linear graph of any data
sns.lineplot(x='var1',y='var2',data=df)
#data is defined from where to take the x and y value for statistical graph.
plt.show()
```

# Advantage of Seaborn 

***It adds additional details on it on.***

```python
df1=sns.load_dataset("penguins")
df1
```

```python
#USING LINEPLOT DRAW GRAPH FOR DATAFRAME
sns.lineplot(x=	"bill_length_mm",y="flipper_length_mm",data=df1)
plt.show()
```

```python
#same graph in categorical data--->using hue parameter
sns.lineplot(x=	"bill_length_mm",y="flipper_length_mm",data=df1,hue="sex")
plt.show()
```

```python
#changing color for the graph--->palette parameter is used
#for making points visible use the marker parameter using style parameter as well
sns.lineplot(x=	"bill_length_mm",y="flipper_length_mm",data=df1,hue="sex",palette="rocket_r",style="sex",markers=["o",">"])
plt.show()
```

```python
#To make grid in your ploted graph 
sns.lineplot(x=	"bill_length_mm",y="flipper_length_mm",data=df1,hue="sex",palette="rocket_r",style="sex",markers=["o",">"])
plt.grid()
plt.show()
```

# BAR PLOT IN SEABORAN LIBRARY

```python
df2=sns.load_dataset("penguins")
df2
```

```python
df2["island"]
```

```python
sns.barplot(x="island",y="bill_length_mm",data=df2,width=0.5)
plt.show()
```

```python
#To get the data in terms of sex for each island 
sns.barplot(x="island",y="bill_length_mm",data=df2,width=0.5,hue="sex")
plt.show()
```

# Order Parameter

```python
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df2,hue="sex",order=order1)
plt.show()
```

```python
# TO CHANGE ORDER OF REPRESENTATION OF BARS IN GRAPHICAL FORM-->hue_order
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df2,hue="sex",order=order1,hue_order=['Female','Male'])
plt.show()
```

```python
#TO SET COLORS ACCORDING TO US-->color parameter
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df2,order=order1,
            color="pink")
plt.show()
```

```python
#TO SET COLORS ACCORDING TO US IN HUE CONCEPT-->palette parameter
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df2,order=order1,hue="sex",
            palette="Accent",hue_order=["Female","Male"])
plt.show()
```

# SATURATION PARAMETER

```python
#SATURAION PARAMETER IS USED FOR DULLING COLORS OR BRIGHTANING THEM.
#0<SATURATION PARAMETER<1
order1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=df2,order=order1,hue="sex",
            palette="Accent",hue_order=["Female","Male"],saturation=0.7)
plt.show()
```

# HISTOGRAM IN SEABORN LIBRARY

```python
#check the dataset
df2
```

```python
sns.histplot(df2["flipper_length_mm"])
plt.show()
```

```python
#In above graph proper intervals are not visible.
#To get proper interval--->bins parameter
sns.displot(df2["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

# KERNEL DENSITY IN HISTOGRAM

```python
#KERNEL DENSITY SHOWS HOW MUCH RAISE IS MADE AND HOW MUCH FALL IS SEEN IN GRAPH
#KDE PARAMETER IS USED TO SEE DENSITY.
sns.displot(df2["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240], kde=True)
plt.show()
```
